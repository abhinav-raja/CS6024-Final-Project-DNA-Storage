```python
import matplotlib.pyplot as plt
```

```python
len = [500, 1000, 2000, 3000]
orig = [0.28, 1.18, 4.73, 10.69]
mod = [0.15, 0.62, 2.38, 5.5]

plt.plot(len, orig, label="Original LCS")
plt.plot(len, mod, label="Kuo Cross")
plt.xlabel("Length of the strings")
plt.ylabel("Time taken (s)")
plt.legend()
plt.title("Varying lengths ; Random pairs")
```
